```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
name = 18
file = pd.read_csv(f"../data-source/4-11-21/S{name}.csv")
```

```python
start = 0
end   = 160

data = file.to_numpy().transpose()
fig, ax = plt.subplots(figsize=(7,5))

ax.plot(data[0][start:end], data[1][start:end], color = "black")
ax.scatter(data[0][start:end], data[1][start:end], color = "black")
ax.grid()
ax.set(xlabel = "Base dei tempi (s)", ylabel = "Codici")
print("Intervallo di campionamento = ", data[0][0] - data[0][1], " secondi")
plt.savefig("./output/letture_mancanti.pdf")
```
